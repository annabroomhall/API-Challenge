```python
import matplotlib.pyplot as plt
import requests
from scipy import stats
import pandas as pd
from config import weather_api_key
from pprint import pprint
from citipy import citipy
import random as rd
import scipy.stats as st
from scipy import stats

url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"

# Build partial query URL
query_url = f"{url}appid={weather_api_key}&units={units}&q="
```

## Generate Cities List

```python
# create a list to store the cities in
city_list = []

# create a list of 1000 cities
while len(city_list) < 1000:
    
    # set random lat and long
    latitude = rd.uniform(-90.0,90.0)
    longitude = rd.uniform(-180.0,180.0)
    
    # get nearest city
    city = citipy.nearest_city(latitude,longitude).city_name
        
    if city not in city_list:
        city_list.append(city)
```

```python
city_list
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# set up lists to hold reponse info
cities = city_list
lat = []
lng = []
max_temp = []
humidity = []
cloudiness = []
wind_speed = []
country = []
date = []
name = []
number = []


# Loop through the list of cities and perform a request for data on each
for city in city_list:
    try:
        response = requests.get(query_url + city).json()
        lat.append(response['coord']['lat'])
        lng.append(response['coord']['lon'])
        max_temp.append(response['main']['temp_max'])
        humidity.append(response['main']['humidity'])
        cloudiness.append(response['clouds']['all'])
        wind_speed.append(response['wind']['speed'])
        country.append(response['sys']['country'])
        date.append(response['dt'])
        name.append(response['name'])
        number.append(response['id'])
        print(f"City: {(response['name'])} (ID: {(response['id'])}) located -- Updating data.")
    except:
        print(f"{city} not found")
     
```

```python
len(lat)
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
city_data = pd.DataFrame({
                                  "City": name,
                                  "Country": country,
                                  "City ID": number,
                                  "Lat": lat,
                                  "Long": lng,
                                  "Max Temp": max_temp,
                                  "Humidity": humidity,
                                  "Cloudiness": cloudiness,
                                  "Wind Speed": wind_speed,
                                  "Date": date
                                 })

city_data
```

```python
city_data_clean = city_data.drop_duplicates(subset=['City'])
city_data_clean = city_data_clean.dropna(how='any')

len(city_data_clean)
```

```python
city_data_clean.to_csv('citydataclean.csv', index=False)
```

## Linear Regression

```python
#Data by Hemisphere

# Create the bins in which Data will be held
bins = [-90, 0, 90]

# Names for the bins
hemisphere = ["Southern Hemisphere", "Northern Hemisphere"]

# Add bins to city_data table
city_data["Hemisphere"] = pd.cut(city_data["Lat"], 
                                          bins, labels=hemisphere, include_lowest=True)

city_data_north = city_data.loc[city_data['Hemisphere'] == 'Northern Hemisphere']
city_data_south = city_data.loc[city_data['Hemisphere'] == 'Southern Hemisphere']
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
# Generate a scatter plot of latitude versus temperature in Northern Hemisphere

#set data for scatter plot
lat = city_data_north.iloc[:,3]
max_temp = city_data_north.iloc[:,5]
plt.scatter(lat,max_temp)

nmt_slope, nmt_int, nmt_r, nmt_p, nmt_std_err = stats.linregress(lat,max_temp)
nmt_regress = lat * nmt_slope + nmt_int

#Plot linear regression
plt.plot(lat,nmt_regress,"r-",linewidth=1)

#Annotate linear regression
line_eq = "y = " + str(round(nmt_slope,2)) + "x + " + str(round(nmt_int,2))
plt.annotate(line_eq,(40,80),fontsize = 15,color = "red")


#Add labels and title to scatter plot
#plt.xticks(max_temp)
plt.ylabel('Max Temperature')
plt.xlabel('Latitude')
plt.title("Northern Hemisphere Temperature vs Latitude on 19/03/2021")
plt.show()
```

There is a strong correlation between the temperature and latitude in the Northern Hemisphere - the weather gets warmer closer to the equator.
